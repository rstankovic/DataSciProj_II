```python
import pandas as pd
import os
```

```python
#/Users/christianstankovic/Desktop/DATA/DataSciProj_II/Project 2 Data/IEA_combined_csv

#values required outside scope of for loop
prevCountry = ''
currCountry = ''
currCSV = ''
currCountryCSV = ''
prevCountryCSVBalances = ''
prevCountryCSVRenewables = ''

#the actual process
for file in sorted(os.listdir('/Users/christianstankovic/Desktop/DATA/DataSciProj_II/Project 2 Data/IEA_CSVs_raw')):
    #values of current file
    currCountry = file.split('_')[0]
    category = file.split('_')[2].split('.')[0]
    year = str(file.split('_')[1])
    
    #seeing if this file is information for the same country as the previous or not
    if year == '2004':
        if file.split('_')[2].split('.')[0] == 'Balances':
            prevCountryCSVBalances = pd.read_csv(f'/Users/christianstankovic/Desktop/DATA/DataSciProj_II/Project 2 Data/IEA_CSVs_raw/{file}')
            prevCountryCSVBalances['Year'] = year #whole point of this is to remove separate files from year to year
            prevCountryCSVBalances.drop(columns = ['Unnamed: 0','Unnamed: 0.1'], inplace = True)
            
        elif file.split('_')[2].split('.')[0] == 'Renewables & waste':
            prevCountryCSVRenewables = pd.read_csv(f'/Users/christianstankovic/Desktop/DATA/DataSciProj_II/Project 2 Data/IEA_CSVs_raw/{file}')
            prevCountryCSVRenewables['Year'] = year #whole point of this is to remove separate files from year to year
            prevCountryCSVRenewables.drop(columns = ['Unnamed: 0','Unnamed: 0.1'], inplace = True)
            
        continue # no csv to combine with it yet as this is the first csv for each column
        
    else: #this is if 
        currCountryCSV = pd.read_csv(f'/Users/christianstankovic/Desktop/DATA/DataSciProj_II/Project 2 Data/IEA_CSVs_raw/{file}')
        currCountryCSV.drop(columns = ['Unnamed: 0','Unnamed: 0.1'], inplace = True)
        currCountryCSV['Year'] = year
        
        if category == 'Balances':
            prevCountryCSVBalances = prevCountryCSVBalances.append(currCountryCSV)
            prevCountryCSVBalances.to_csv(f'/Users/christianstankovic/Desktop/DATA/DataSciProj_II/Project 2 Data/IEA_combined_csv/{currCountry}_{category}.csv')
            
        elif category == 'Renewables & waste':
            prevCountryCSVRenewables = prevCountryCSVRenewables.append(currCountryCSV)
            prevCountryCSVRenewables.to_csv(f'/Users/christianstankovic/Desktop/DATA/DataSciProj_II/Project 2 Data/IEA_combined_csv/{currCountry}_{category}.csv')
            
            
```
