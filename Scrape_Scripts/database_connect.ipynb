```python
for file in countryFiles:
    country_df = pd.read_csv(f'{directory_path}/{file}',index_col = False)
    country_df['Country'] = file.split('_')[0]
    country_df['Year'] = country_df['Year'].apply(str)
    pd.to_numeric(country_df['Year']) 
    country_df.drop(columns = ['Unnamed: 0'], inplace = True)
    category = file.split('_')[1].split('.')[0]
    country = file.split('_')[0]
#     collection.update_one({'_id': url},  {"$set": data}, upsert=True)

    if category == 'Balances':
        for x in range(0, country_df['Year'].size - 1):
            db.Balances.insert_one(country_df.iloc[x].to_dict())
    elif category == 'Renewables & waste':
        for x in range(0, country_df['Year'].size - 1):
            db.Renewables.insert_one(country_df.iloc[x].to_dict())
    print(f"\n----------------- {country} complete -------------------\n")
```
